```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Parámetros del sistema
m = 1.0  # masa (kg)
k = 1.0  # constante del resorte (N/m)
A = 1.0  # Amplitud de la oscilación (m)
omega = np.sqrt(k/m)  # Frecuencia angular (rad/s)

# Definir el tiempo de simulación
t = np.linspace(0, 10, 500)  # 10 segundos, 500 puntos

# Calcular posición y velocidad
x = A * np.cos(omega * t)
v = -A * omega * np.sin(omega * t)

# Calcular energías
U = 0.5 * k * x**2
K = 0.5 * m * v**2
E = U + K  # Debe ser constante si la energía se conserva
```

```python
# Mostrar parámetros con Markdown
display(Markdown(f"### Parámetros del Sistema"))
display(Markdown(f"- Masa, $m$: {m} kg"))
display(Markdown(f"- Constante del resorte, $k$: {k} N/m"))
display(Markdown(f"- Amplitud, $A$: {A} m"))
display(Markdown(f"- Frecuencia angular, $\\omega$: {omega:.2f} rad/s"))
display(Markdown(f"### Energía media calculada"))
display(Markdown(f"- Energía total media (debería ser constante): {np.mean(E):.2f} J"))
```

```python
# Verificar la conservación de la energía
print("Energía total (debería ser constante):", np.mean(E))

### Paso 2: Graficar las energías

plt.figure(figsize=(10, 5))
plt.plot(t, U, label='Energía Potencial (U)', color='red')
plt.plot(t, K, label='Energía Cinética (K)', color='blue')
plt.plot(t, E, label='Energía Total (E)', color='green', linestyle='--')
plt.xlabel('Tiempo (s)')
plt.ylabel('Energía (Joules)')
plt.title('Energía en un Oscilador Armónico Simple')
plt.legend()
plt.grid(True)
plt.show()
```
