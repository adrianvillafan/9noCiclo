```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Parámetros del sistema
m = 1.0  # masa (kg)
k = 1.0  # constante del resorte (N/m)
A = 1.0  # Amplitud de la oscilación (m)
omega = np.sqrt(k/m)  # Frecuencia angular (rad/s)

# Crear un array de posiciones x desde -A hasta A
x = np.linspace(-A, A, 500)

# Calcular las energías en función de x
U = 0.5 * k * x**2
K = 0.5 * k * A**2 - U  # La energía cinética es la energía total menos la energía potencial

# Energía total, que debería ser constante
E = 0.5 * k * A**2
```

```python
# Mostrar parámetros con Markdown
display(Markdown(f"### Parámetros del Sistema"))
display(Markdown(f"- Masa, $m$: {m} kg"))
display(Markdown(f"- Constante del resorte, $k$: {k} N/m"))
display(Markdown(f"- Amplitud, $A$: {A} m"))
display(Markdown(f"- Frecuencia angular, $\\omega$: {omega:.2f} rad/s"))
display(Markdown(f"### Energía total calculada"))
display(Markdown(f"- Energía total constante: {E:.2f} J"))
```

```python
# Graficar las energías en función de la posición x
plt.figure(figsize=(10, 5))
plt.plot(x, U, label='Energía Potencial (U)', color='red')
plt.plot(x, K, label='Energía Cinética (K)', color='blue')
plt.plot(x, [E] * len(x), label='Energía Total (E)', linestyle='--', color='green')
plt.xlabel('Posición x (m)')
plt.ylabel('Energía (Joules)')
plt.title('Energía en función de la Posición en un Oscilador Armónico Simple')
plt.legend()
plt.grid(True)
plt.show()
```
